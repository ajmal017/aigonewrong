```python
import numpy as np
import pandas as pd
import yfinance as yf

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# get historical daily price for SPY
symbol = 'SPY'
tick = yf.Ticker(symbol)
history = tick.history(period="max")
# refresher https://www.portfolioprobe.com/2010/10/04/a-tale-of-two-returns/
# https://quant.stackexchange.com/questions/21092/calculating-log-returns-across-multiple-securities-and-time
#history['log_ret'] = np.log(history.Close) - np.log(history.Close.shift(1))
#history = history.dropna()
close = history.Close
log_ret = np.diff(np.log(close))
```

```python
close_reversed = np.cumprod(log_ret+1)*close[0]
plt.plot(close_reversed)
plt.plot(close.values)
plt.grid(True)
print(np.cumprod(log_ret+1)[-1],close_reversed[-1]/close_reversed[0])
```
